```python
import pandas as pd
```

```python
sabio = pd.read_csv('sabio_df_checkpoint2.csv')
glucose_cpd = 'C00031'
phosphofructokinase_1_ko = ''
```

```python
only_gluc_sabio = []
for ind, entry in enumerate(sabio['KEGG ID']):
    if entry == glucose_cpd:
        only_gluc_sabio.append(ind)
only_gluc_sabio = sabio.take(only_gluc_sabio, axis=0)
```

```python
sabio_lac_dh = pd.read_table('sabio_lactate_dh.tsv')
lac_dh_grouped = sabio_lac_dh.groupby('EntryID')
```

```python
dataframes = []
for entry in lac_dh_grouped:
    for ind, param in enumerate(entry[1]['parameter.name']):
        if param == 'Km':
            dataframes.append(entry[1].iloc[ind])
```

```python
dataframes[0]
```

```python
refined_lac_dh = pd.DataFrame(dataframes).drop(['parameter.endValue', 'parameter.name', 'parameter.type'], inplace=False, axis=1)
refined_lac_dh = refined_lac_dh.rename({'parameter.associatedSpecies' : 'Compound', 'parameter.startValue' : 'Km', 'parameter.unit' : 'unit'})
refined_lac_dh.reset_index(inplace=True)
refined_lac_dh.drop(['index'], axis=1, inplace=True)
refined_lac_dh
```
